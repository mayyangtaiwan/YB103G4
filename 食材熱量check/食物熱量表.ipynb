```python
#爬每個類別的links, save到links.txt
import requests
from bs4 import BeautifulSoup
import string
links = open('links.txt', 'w')

c = 'https://web.archive.org/web/20090330102612/http://tw.health.yahoo.com/030430/26/xse.html'
print c
links.write(c+'\n')
for m in range (2, 3):
    d = 'https://web.archive.org/web/20090330102612/http://tw.health.yahoo.com/030430/26/xse'+str(m)+'.html'
    print d
    links.write(d+'\n')

res = requests.get(c)
soup = BeautifulSoup(res.text.encode('utf8'))
x = soup.select('.sbody')
#print x
for i in x:
    j = i.findAll('a', {'href':True})
    for k in j:
        if 'tw.health.yahoo.com' in k['href']:
            a = 'https://web.archive.org'+k['href']
            #print a
            links.write(a+'\n')
        for l in range(2, 5):
            if 'tw.health.yahoo.com' in k['href']:
                b = 'https://web.archive.org'+k['href'].split('.html')[0]+str(l)+'.html'
                #print b
                links.write(b+'\n')

links.close()
```

```python
import requests
from bs4 import BeautifulSoup
Hey = open('creation.txt', 'w')
page_format= ('http://www.creation.com.tw/product/content.asp?bid=%d')
TitleLs = []
for page in range(10, 333):
    res = requests.get(page_format%(page))
    res.encoding = 'utf-8'
    soup = BeautifulSoup(res.text.encode('utf-8')) 
    try:
        Title = soup.select('strong')[0].text.strip()
        TitleLs.append(Title)    
        Qaa = soup.select('table h4 span')[0].text.replace('(', '').replace(')', '').strip()
        TitleLs.append(Qaa)    
        Qbb = soup.select('table td')[5].text.strip()
        TitleLs.append(Qbb)
    except: pass
for i in TitleLs:
    Hey.write((i+",").encode('utf-8'))
Hey.write("\n")
Hey.close()
```

```python
import requests
from bs4 import BeautifulSoup
import string
kk = open('test.txt', 'w')
res = requests.get('http://www.dreye-health.com/SingularController?service=CalorieList#')
soup = BeautifulSoup(res.text.encode('utf-8'))
food = soup.findAll('a', {'onclick': True})
for i in food:
    j = i['onclick']
    if 'return addFoodList' in j:
        k = j.split(',')
        for qoo in k[1:5]:
            kk.write(qoo.encode('utf-8'))
        kk.write('\n')
kk.close()
```

```python
#myfitnesspal
import requests
from bs4 import BeautifulSoup
myfitnesspal = open('myfitnesspal.txt', 'w')
page_format= ('http://www.myfitnesspal.com.tw/food/update_nutrition_facts/%d')
for page in range(100000000, 200000000):
    res = requests.get(page_format%(page))
    soup = BeautifulSoup(res.text) 
    try:
        if soup.select('.food-description')[0].text is not None:
            Qaa = soup.select('.food-description')[0].text
            Qbb = soup.select('select#food_entry_weight_id option')[0].text
            Qcc = soup.select('#nutrition-facts .col-2')[2].text
            myfitnesspal.write(Qaa+','+Qbb+','+Qcc+'\n')
    except:
        pass
myfitnesspal.close()
```
