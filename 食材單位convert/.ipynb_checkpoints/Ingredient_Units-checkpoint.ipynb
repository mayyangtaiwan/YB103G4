```python
in_file = open('0421_unit_counts_with_cat.csv','r')
out_file = open('All-units.txt','w')

def col(line, n):
    while (line.split(',')[n] is not '') and (line.split(',')[n+1] is not ''):
        Qaa = line.split(',')[0].strip()+','+line.split(',')[1].strip()+','
        #print line.split(',')[0], line.split(',')[1], line.split(',')[n+1]
        out_file.write(Qaa+line.split(',')[n+1].strip()+','+''+'\n')
        n += 2
    return

for line in in_file.readlines():
    col(line, 2)   
        
in_file.close()
out_file.close()


#remove 單位是數字 及 單位有括號 的rows
import re
aaa = open('All-units.txt','r')
bbb = open('All-units-clean.txt','w')
for i in aaa.readlines():
    ii = i.split(',')[2]
    m = re.match(r'(\d+)', ii)
    if m:
        pass
    else:
        if '(' in ii:
            iii = ii.split('(')[0]
            bbb.write(i.split(',')[0]+','+i.split(',')[1]+','+iii+',\n')
        elif '（' in ii:
            jjj = ii.split('（')[0]
            bbb.write(i.split(',')[0]+','+i.split(',')[1]+','+jjj+',\n')
        else:
            bbb.write(i)

aaa.close()
bbb.close()
```
