```python
import requests, time
from bs4 import BeautifulSoup

fooding_file = open("fooding_file.txt","r")                # 讀檔
fooding_ingregients = open("fooding_ingregients.txt","w")  # 寫入
fooding_practice= open("fooding_practice.txt","w")         # 寫入

for line in fooding_file.readlines():                      # 讀fooding_file的每一行
    #try:
        pagelink = line.strip()
        #print pagelink
        res = requests.get(pagelink)
        res.encoding ='utf-8'
        soup = BeautifulSoup(res.text)
        
        # 食譜名稱
        title = soup.select(".m3")[0].select('h1')[0].text 
        print title
            
        # 食材
        ingredients = soup.select(".info-right")[0].text 
        print ingredients
        
        # 作法
        practice  = soup.select(".mod_con")[0].text 
        #for step in practice:
        print practice
        
        
        fooding_ingregients.write(title.encode('utf-8')+',') # 標題
        for item  in ingredients:                                 # 食材
             fooding_ingregients.write("%s, %s,"%(item('.info-right')[0].text.encode('utf-8')))
        fooding_ingregients.write('\n')
        
        
        fooding_practice.write(title.text.encode('utf-8')+',')    # 標題 
        for step  in practice:                                    # 作法
             fooding_practice.write(step.text.encode('utf-8'))
        fooding_practice.write('\n')
        
        fooding_ingregients.write((title+',').encode('utf-8')) # 標題 write((i+',').encode('utf-8'))
        for item  in ingredients:                                 # 食材
             fooding_ingregients.write("%s, %s,"%(item('.info-right')[0].text.encode('utf-8')))
        fooding_ingregients.write('\n')
        
        
        fooding_practice.write(title.text.encode('utf-8')+',')    # 標題 
        for step  in practice:                                    # 作法
             fooding_practice.write(step.text.encode('utf-8'))
        fooding_practice.write('\n')
        
        
    #except:
        #pass
      
    
fooding_practice.close()   
fooding_ingregients.close() 

time.sleep(1)

fooding_file.close()
'''

TypeError                                 Traceback (most recent call last)<ipython-input-6-e690d76bf09c> in <module>()     
30         fooding_ingregients.write(title.encode('utf-8')+',') # 標題     
31         for item  in ingredients:                                 # 食材---> 
32              fooding_ingregients.write("%s, %s,"%(item('.info-right')[0].text.encode('utf-8')))     
33         fooding_ingregients.write('\n')     
34 
TypeError: 'unicode' object is not callable  

Q : TypeError: 'unicode' object is not callable  → 有轉成 utf-8 還是error 要怎麼修改呢?
'''
```
