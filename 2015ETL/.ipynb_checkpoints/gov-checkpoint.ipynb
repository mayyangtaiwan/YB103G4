```python
#行政院農委會 All

import requests
from bs4 import BeautifulSoup
import re
homepage = requests.get('http://www.coa.gov.tw/view.php?catid=303')
homepage.encoding = 'big5'
homepageSoup = BeautifulSoup(homepage.text.encode('utf-8'))
homepageTable  = homepageSoup.select(".context_COA")[0].select("a")

#爬出第一層
gov_recipe_file = open("gov_recipe_list.txt", 'w')
for tag in homepageTable:
    if re.match("(^view*)", tag['href']):
        gov_recipe_file.write("http://www.coa.gov.tw/" + tag['href'] + "\n") 
gov_recipe_file.close()

#爬出第二層
gov_recipe_file = open("gov_recipe_list.txt", 'r')
for line in gov_recipe_file.readlines(): 
    pageTwoLink = line.strip() 
    pageTwo_get = requests.get(pageTwoLink)
    pageTwo_get.encoding = 'big5'
    pageTwoSoup = BeautifulSoup(pageTwo_get.text.encode('utf-8'))
    pageTwoTable  = pageTwoSoup.select(".context_COA")[0].select("a")
    #print pageTwoTable
    pageTwo_file = open("pageTwo_list.txt", 'a')
    for tag in pageTwoTable:
        if re.match("(^view*)", tag['href']):
            pageTwo_file.write("http://www.coa.gov.tw/" + tag['href'] + "\n") 
pageTwo_file.close() 
gov_recipe_file.close()

#爬出第三層
pageTwo_recipe_file = open("pageTwo_list.txt", 'r')
for line in pageTwo_recipe_file.readlines(): 
    pageThreeLink = line.strip() 
    pageThree_get = requests.get(pageThreeLink)
    pageThree_get.encoding = 'big5'
    pageThreeSoup = BeautifulSoup(pageThree_get.text.encode('utf-8'))
       
    if pageThreeSoup.find("h3") is None:
        pageThreeTable  = pageThreeSoup.select(".context_COA")[0].select("a") 
        pageThree_file = open("pageThree_list.txt", 'a')
        for tag in pageThreeTable:
            if re.match("(^view*)", tag['href']):
                pageThree_file.write("http://www.coa.gov.tw/" + tag['href'] + "\n") 
        
    elif pageThreeSoup.find("h3").text.encode('utf-8').strip() == "材料":
        gov_ingredients = open('gov_ingredients.txt','a')
        gov_steps = open('gov_steps.txt','a')

        try:             
            ingredientsList = []
            title = pageThreeSoup.select('.h4')[0].text.encode('utf-8')
            ingredientsList.append(title)
            ingredients = pageThreeSoup.select('pre')[0].text.encode('utf-8').strip()
            ingredientsList.append(ingredients)
            seasoning1 = pageThreeSoup.select('pre')[1].text.encode('utf-8').strip()
            ingredientsList.append(seasoning1) 
            #seasoning2 = pageThreeSoup.select('pre')[1].select('p')[0].text.encode('utf-8').strip()
            #ingredientsList.append(seasoning2)

            stepsList = []
            stepsList.append(title)
            stepsList.append(pageThreeSoup.select('ol')[0].text.encode('utf-8').strip())
            
            for i in ingredientsList:
                gov_ingredients.write((i+','))
            gov_ingredients.write('\n')    
            for i in stepsList:
                gov_steps.write((i+','))
            gov_steps.write('\n') 

        except:            
            pass   
        
pageThree_file.close()
gov_ingredients.close()
gov_steps.close()
pageTwo_recipe_file.close()

#爬出第四層
pageThree_recipe_file = open("pageThree_list.txt", 'r')
for line in pageThree_recipe_file.readlines(): 
    pageFourLink = line.strip() 
    pageFour_get = requests.get(pageFourLink)
    pageFour_get.encoding = 'big5'
    pageFourSoup = BeautifulSoup(pageFour_get.text.encode('utf-8'))
        
    if pageFourSoup.find("h3") is None:
        pageFourTable  = pageFourSoup.select(".context_COA")[0].select("a") 
        pageFour_file = open("pageFour_list.txt", 'a')
        for tag in pageFourTable:
            if re.match("(^view*)", tag['href']):
                pageFour_file.write("http://www.coa.gov.tw/" + tag['href'] + "\n") 

    elif pageFourSoup.find("h3").text.encode('utf-8').strip() == "材料":
        gov_ingredients = open('gov_ingredients.txt','a')
        gov_steps = open('gov_steps.txt','a')
        
        try:           
            ingredientsList = []
            title = pageFourSoup.select('.h4')[0].text.encode('utf-8')
            ingredientsList.append(title)
            ingredients = pageFourSoup.select('pre')[0].text.encode('utf-8').strip()
            ingredientsList.append(ingredients)
            seasoning1 = pageFourSoup.select('pre')[1].text.encode('utf-8').strip()
            ingredientsList.append(seasoning1)        
            #seasoning2 = pageFourSoup.select('pre')[1].select('p')[0].text.encode('utf-8').strip()
            #ingredientsList.append(seasoning2)

            stepsList = []
            stepsList.append(title)
            stepsList.append(pageFourSoup.select('ol')[0].text.encode('utf-8').strip())
            
            for i in ingredientsList:
                gov_ingredients.write((i+','))
            gov_ingredients.write('\n')    
            for i in stepsList:
                gov_steps.write((i+','))
            gov_steps.write('\n')
 
        except:
            pass  
                
pageFour_file.close()
gov_steps.close()
gov_ingredients.close()                    
pageThree_recipe_file.close()

#爬出第五層
pageFour_recipe_file = open("pageFour_list.txt", 'r')
for line in pageFour_recipe_file.readlines(): 
    pageFiveLink = line.strip() 
    pageFive_get = requests.get(pageFiveLink)
    pageFive_get.encoding = 'big5'
    pageFiveSoup = BeautifulSoup(pageFive_get.text.encode('utf-8'))
        
    if pageFiveSoup.find("h3") is None:
        pageFiveTable  = pageFiveSoup.select(".context_COA")[0].select("a") 
        pageFive_file = open("pageFive_list.txt", 'a')
        for tag in pageFiveTable:
            if re.match("(^view*)", tag['href']):
                pageFive_file.write("http://www.coa.gov.tw/" + tag['href'] + "\n") 

    elif pageFiveSoup.find("h3").text.encode('utf-8').strip() == "材料":
        gov_ingredients = open('gov_ingredients.txt','a')
        gov_steps = open('gov_steps.txt','a')
        
        try:           
            ingredientsList = []
            title = pageFiveSoup.select('.h4')[0].text.encode('utf-8')
            ingredientsList.append(title)
            ingredients = pageFiveSoup.select('pre')[0].text.encode('utf-8').strip()
            ingredientsList.append(ingredients)
            seasoning1 = pageFiveSoup.select('pre')[1].text.encode('utf-8').strip()
            ingredientsList.append(seasoning1)        
            #seasoning2 = pageFiveSoup.select('pre')[1].select('p')[0].text.encode('utf-8').strip()
            #ingredientsList.append(seasoning2)

            stepsList = []
            stepsList.append(title)
            stepsList.append(pageFiveSoup.select('ol')[0].text.encode('utf-8').strip())
            
            for i in ingredientsList:
                gov_ingredients.write((i+','))
            gov_ingredients.write('\n')    
            for i in stepsList:
                gov_steps.write((i+','))
            gov_steps.write('\n')

        except:
            pass  
                
pageFive_file.close()
gov_steps.close()
gov_ingredients.close()                    
pageFour_recipe_file.close()

#爬出第六層
pageFive_recipe_file = open("pageFive_list.txt", 'r')
for line in pageFive_recipe_file.readlines(): 
    pageSixLink = line.strip() 
    pageSix_get = requests.get(pageSixLink)
    pageSix_get.encoding = 'big5'
    pageSixSoup = BeautifulSoup(pageSix_get.text.encode('utf-8'))
        
    if pageSixSoup.find("h3") is None:
        pageSixTable  = pageSixSoup.select(".context_COA")[0].select("a") 
        pageSix_file = open("pageSix_list.txt", 'a')
        for tag in pageSixTable:
            if re.match("(^view*)", tag['href']):
                pageSix_file.write("http://www.coa.gov.tw/" + tag['href'] + "\n") 

    elif pageSixSoup.find("h3").text.encode('utf-8').strip() == "材料":
        gov_ingredients = open('gov_ingredients.txt','a')
        gov_steps = open('gov_steps.txt','a')
        
        try:           
            ingredientsList = []
            title = pageSixSoup.select('.h4')[0].text.encode('utf-8')
            ingredientsList.append(title)
            ingredients = pageSixSoup.select('pre')[0].text.encode('utf-8').strip()
            ingredientsList.append(ingredients)
            seasoning1 = pageSixSoup.select('pre')[1].text.encode('utf-8').strip()
            ingredientsList.append(seasoning1)        
            #seasoning2 = pageSixSoup.select('pre')[1].select('p')[0].text.encode('utf-8').strip()
            #ingredientsList.append(seasoning2)

            stepsList = []
            stepsList.append(title)
            stepsList.append(pageSixSoup.select('ol')[0].text.encode('utf-8').strip())
            
            for i in ingredientsList:
                gov_ingredients.write((i+','))
            gov_ingredients.write('\n')    
            for i in stepsList:
                gov_steps.write((i+','))
            gov_steps.write('\n')
 
        except:
            pass  
                
pageSix_file.close()
gov_steps.close()
gov_ingredients.close()                    
pageFive_recipe_file.close()
```
