```python
import os

def parse_file(datafile):
    data = []
    with open(datafile, "rb") as f:
        header = f.readline().split(",")
        #print header
        count = 0
        for line in f:
            if count == 10:
                break
                
            ele = line.split(",")
            entry ={}
            #print ele
            
            for idx, item in enumerate(ele):
                entry[header[idx].strip()]=item.strip()
            #print entry #put together
            
            data.append(entry)
            count += 1
            
    return data

DATADIR = ""
DATAFILE = "beatles-diskography.csv"
parse_file(DATAFILE)
```
